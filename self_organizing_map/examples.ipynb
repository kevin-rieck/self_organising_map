```python
from self_organizing_map.objects import SOM
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(42)
colors = np.random.rand(30, 3)
```

```python
som = SOM(25, 50, dim=3, n_iterations=3000, alpha=1.9, metric='euclidean')
```

```python
som.fit(colors)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(som.cluster)
ax[0].set_title('Aggl. Clustering')
ax[1].imshow(som.get_umatrix())
ax[1].set_title('U-matrix')
ax[2].imshow(som.get_centroids(), interpolation='bicubic')
ax[2].set_title('Colors of centroids')
plt.show()
```

```python
for i in range(100):
    test_vectors = np.random.rand(1000, 3)
    predictions = som.predict(test_vectors, learn_qe=True)
```
